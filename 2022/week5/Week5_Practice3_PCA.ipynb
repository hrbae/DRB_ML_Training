# 1. Load data

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/hrbae/DRB_ML_Training/main/2022/week5/cereals.csv')
print(df.columns)
```

# 2. Data Preprocessing

```python
df = df[['calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups',
       'rating']]

print(df.columns)
```

```python
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale)
df_scale
```

# 3. PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca_x = pca.fit_transform(df_scale)
pca.explained_variance_ratio_
```

```python
pca_rlt = pd.DataFrame(pca_x, columns = ['PC1', 'PC2'])
pca_rlt
```

## Default

```python
pca = PCA()
pca_x = pca.fit_transform(df_scale)
ratio = pca.explained_variance_ratio_
ratio
```

```python
pca_rlt = pd.DataFrame(pca_x, columns = [f'PC{i+1}' for i in range(len(ratio))])
pca_rlt
```

```python
df_ratio = pd.DataFrame(ratio, index = [f'PC{i+1}' for i in range(len(ratio))], columns = ['ratio'] )
df_ratio.plot.pie(y='ratio')
```
