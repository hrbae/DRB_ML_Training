# 1. Load

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
import pandas as pd

df = pd.DataFrame(data=boston.data, columns =boston.feature_names)
df['target'] = boston.target

df.head()
```

# 2. EDA

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

```python
for i in df.columns:
  sns.histplot(df[i])
  plt.show()
```

## Correlation

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(corr)
```
