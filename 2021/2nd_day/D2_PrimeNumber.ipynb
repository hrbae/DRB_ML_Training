** 에라토스테네스의 체**
image.png

```python
no = int(input("어느 범위에서 소수를 구하고 싶은가요?"))

l_no = list(range(2, no+1))


for k in range(0, no-1):
    i = k+2 #i는 2부터 체크

    inc=2

    temp=i*inc

    while temp <= no:
        if temp in l_no:
            l_no.remove(temp)
        inc=inc+1
        temp=i*inc

print("찾아낸 소소:",len(l_no),"개", l_no)
```
