## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
import pandas as pd
df=pd.read_csv('auto.csv')
```

```python
y=df['mpg']
X=df.drop(columns='mpg')
```

```python
from sklearn.model_selection import KFold, cross_validate
kfold=KFold(n_splits=5, shuffle=True)
for i in range(2,30):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X,y)
    result=cross_validate(X=X,y=y,cv=kfold,scoring=['r2','neg_mean_squared_error'],estimator=knn)
    print(i, result['test_r2'].mean(), result['test_neg_mean_squared_error'].mean())
    
    
```

## Impact of Data Scaling

## z-normalization

```python
X_norm = (X-X.mean())/X.std()
```

```python
from sklearn.model_selection import KFold, cross_validate
kfold=KFold(n_splits=5, shuffle=True)
for i in range(2,30):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X,y)
    result=cross_validate(X=X_norm,y=y,cv=kfold,scoring=['r2','neg_mean_squared_error'],estimator=knn)
    print(i, result['test_r2'].mean(), result['test_neg_mean_squared_error'].mean())
    
```

## min-max scaling

```python
X_scale = (X-X.min())/(X.max()-X.min())
```

```python
from sklearn.model_selection import KFold, cross_validate
kfold=KFold(n_splits=5, shuffle=True)
for i in range(2,30):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(X_scale,y)
    result=cross_validate(X=X_norm,y=y,cv=kfold,scoring=['r2','neg_mean_squared_error'],estimator=knn)
    print(i, result['test_r2'].mean(), result['test_neg_mean_squared_error'].mean())
```

## KNN Classification

```python
df = pd.read_csv('heart.csv')
df=df.dropna()
```

```python
df=pd.get_dummies(df,columns=['Sex','AHD','ChestPain','Thal','AHD'])
```

```python
y=df['AHD_Yes']
X=df.drop(columns=['AHD_No','AHD_Yes'])
```

```python
from sklearn.model_selection import KFold, cross_validate
from sklearn.neighbors import KNeighborsClassifier
kfold=KFold(n_splits=5, shuffle=True)
for i in range(2,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X,y)
    result=cross_validate(X=X,y=y,cv=kfold,scoring=['accuracy','roc_auc'],estimator=knn)
    print(i, result['test_accuracy'].mean(), result['test_roc_auc'].mean())
```

## After Normalizing

```python
X_norm=(X-X.mean())/X.std()
X_norm
```

```python
from sklearn.model_selection import KFold, cross_validate
from sklearn.neighbors import KNeighborsClassifier
kfold=KFold(n_splits=5, shuffle=True)
for i in range(2,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_norm,y)
    result=cross_validate(X=X_norm,y=y,cv=kfold,scoring=['accuracy','roc_auc'],estimator=knn)
    print(i, result['test_accuracy'].mean(), result['test_roc_auc'].mean())
```
